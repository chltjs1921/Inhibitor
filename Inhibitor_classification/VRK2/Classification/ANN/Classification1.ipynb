```python
#ANN 정해진 파라미터로 실험

import numpy as np
import pandas as pd
import joblib

from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.layers import Dropout
from keras.constraints import maxnorm
from sklearn.metrics import accuracy_score, confusion_matrix
from keras.models import load_model
from keras.optimizers import Adadelta
from scipy.stats import randint, uniform, loguniform


# fix random seed for reproducibility

seed = 7
np.random.seed(seed)

#데이터 입력
df = pd.read_csv('./VRK2_inhibitor_descriptor_Internal_dataset_dropunmane.csv')
df1 = pd.read_csv('./VRK2_inhibitor_descriptor_external_dataset_dropunmane.csv')
df2 = pd.concat([df, df1])

#데이터 분류
dataset = df.values
X = dataset[:,0:-1]
Y = dataset[:,-1]

ex_data =df1.values
exX = ex_data[:,0:-1]
exY = ex_data[:,-1]

tot_data = df2.values
totX = tot_data[:,0:-1]
totY = tot_data[:,-1]

#X 표준화

scaler = StandardScaler()

scaler.fit(X)
X_scaled = scaler.transform(X)
exX_scaled = scaler.transform(exX)
totX_scaled = scaler.transform(totX)

#모델 설정

n_feat = X_scaled.shape[1]
n_class = len(set(Y))

model = Sequential()
model.add(Dense(290, input_dim=n_feat , kernel_initializer='lecun_uniform', activation='sigmoid', kernel_constraint=maxnorm(5)))
model.add(Dense(n_class, kernel_initializer='lecun_uniform', activation='softmax'))
optimizer = Adadelta(lr=0.001)
model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

#model.summary()


#모델 학습
model.fit(X_scaled,Y, epochs=79, batch_size=90)

#학습 평가

print(model.evaluate(X_scaled, Y)[1])
print(model.evaluate(exX_scaled, exY)[1])
print(model.evaluate(totX_scaled, totY)[1])
```
