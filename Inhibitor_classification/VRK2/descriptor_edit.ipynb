```python
import pandas as pd

from sklearn.feature_selection import SelectKBest, f_classif
from google.colab import drive
drive.mount('/content/drive')

#데이터 입력
df = pd.read_csv('/content/drive/MyDrive/Inhibitor classification/Inhibitor/VRK2/MORDRED(2D)/VRK2_mordred(2D)_apppendscore_dropsmiles.csv')

#df_info = df.describe()

#df_info.to_csv('/content/drive/MyDrive/Inhibitor classification/Inhibitor/VRK2/MORDRED(2D)/VRK2_mordred(2D)_info.csv') #모든 칼럼이 0인 칼럼을 찾아내기 위해 칼럼들 정보 추출. 추출된 정보를 열어서 최대와 최소 모두 0인 임의의 칼럼 찾아내기

#데이터 분류
dataset = df.values
X = dataset[:,0:-1]
Y = dataset[:,-1]

#피처 엔지니어링

selectK = SelectKBest(score_func=f_classif, k=100)
X = selectK.fit_transform(X, Y)

print(X.shape)

pd.DataFrame(X).to_csv('/content/drive/MyDrive/Inhibitor classification/Inhibitor/VRK2/MORDRED(2D)/VRK2_mordred(2D)_apppendscore_dropsmiles_100feat.csv',index=False) 
```
